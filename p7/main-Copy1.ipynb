```python
import csv

fifa_file = open('Fifa19.csv', encoding='utf-8')
file_reader = csv.reader(fifa_file)
player_data = list(file_reader)
fifa_file.close()
header = player_data[0]
player_data = player_data[1:]
for row in player_data:
    for idx in [2,4]:
        row[idx] = float(row[idx])
```

```python
def cell(row_idx, col_name):
    col_idx = header.index(col_name)
    val = player_data[row_idx][col_idx]
    if val == "":
        return None
    if col_name == 'Id':
        val = float(val)
        return val
    if col_name == 'Age':
        val = int(val)
        return val
    if col_name == 'Overall':
        val = int(val)
        return val
    if col_name == 'Jersey Number':
        val = float(val)
        return val
    if col_name == 'Value':
        if val[-1] == 'M':
            return int(float(val[1:-1])*1000000)
        elif val[-1] == 'K':
            return int(float(val[1:-1])*1000)
        else:
            return int(val[1:])
    if col_name == 'Wage':
        val = int(float(val.replace('K','')[1:])*1000)
    return val
```

```python
#q18
def sum_overall():
    buckets = {}
    for row in player_data:
        jn = row[header.index("Jersey Number")]
        overall = row[header.index("Overall")]
        if not jn in buckets:
            buckets[jn] = overall
        else:
            buckets[jn] += overall
    return buckets

num_overall = bucketize("Jersey Number")
sum_overall_dict = sum_overall()
result = {}
for key in sum_overall_dict:
    result[key] = sum_overall_dict[key] / num_overall[key]
result
```
